## Introduction

This project revolves around the process of parsing an XML file into a structured DataFrame, followed by thorough data auditing and cleaning. The primary challenge involved in this integration task was the initial parsing of the XML data using the BeautifulSoup library. During this phase, a strategy was implemented to assign a null value to the 'Source' attribute if a record was not encapsulated within a 'Source' tag to ensure that any records lacking a source were handled properly.

After reviewing the XML file's structure, the data was parsed and transformed into a pandas DataFrame. One noteworthy issue encountered was the formatting of date values in the YYYYMMDDTHHMMSS format. To address this, the datetime library was used to convert these dates into the desired YYYY-MM-DD HH:MM:SS format. Some dates caused an error because they were in the YYYYDDMMTHHMMSS format, so conditional statements were used for proper conversion.

The next phase focused on data cleansing. Each column in the DataFrame was checked against the specifications provided in the assignment description. The 'astype' method was used in assigning their appropriate data types. 'value_counts' helped in identifying possible misspelled words within the data. Challenges were encountered during the conversion of the 'Salary' column to float due to the presence of strings. To resolve this, 'k'-appended values were modified to reflect thousands ('000'). Range representations such as "to," "-", "~," and "" were resolved by preserving the higher value. Similarly, strings like "per year," "/year," and "per Annum" were removed from 'Salary' entries, leaving only the numeric value. Following these adjustments, the 'Salary' column was successfully transformed into float format.

To address missing values, various placeholders like "N/A," "None," "", and "-" were replaced by "non-specified" across the columns in mentioned in the assignment description. The detection of outliers in the Salary column were performed using a box plot, with identified outliers being subsequently removed from the DataFrame. The integrity constraints of the data were upheld by verifying that 'OpenDate' values preceded corresponding 'CloseDate' values. When discrepancies were identified, 'OpenDate' and 'CloseDate' values were switched to ensure data consistency.

Lastly, the 'Source' column was subjected to a consistency check. While the majority of entries represented websites, a subset were just company names. To rectify this inconsistency, websites were obtained from the internet and matched for these entries, effectively aligning the data.

## Parsing Data

### Examining and loading data

```python
# Code to inspect the provided data file
# reading the XML file using BeautifulSoup and  inspecting its structure using prettify
# Source tag is encapsulating multiple records
# The columns are encapsulated by the record tag
# opendate and closedate encapsulated by date tag
with open("dataset1.xml", "r") as file:
    xml_content = file.read()
soup = BeautifulSoup(xml_content, "lxml-xml")
print(soup.prettify()[:3000])  # Print the first 3000 characters only
```

```python
# Creating Errors Dataframe
data = {
    'indexOfdf': [],
    'Id': [],
    'ColumnName': [],
    'Original': [],
    'Modified': [],
    'ErrorType': [],
    'Fixing': [],
}
errors = pd.DataFrame(data)
```

```python
#Parse data
data = []
current_source = None  #keep track of source

for record in soup.find_all("Record"):
    source = record.find_parent("Source")
    
    # Check if the current record has a source
    if source:
        current_source = source["Name"]
    else:
        current_source = None
    
    record_data = {
        "Id": record.find("Id").get_text(strip=True) if record.find("Id") else None,
        "Title": record.find("Title").get_text(strip=True) if record.find("Title") else None,
        "Location": record.find("Location").get_text(strip=True) if record.find("Location") else None,
        "Company": record.find("Company").get_text(strip=True) if record.find("Company") else None,
        "ContractType": record.find("ContractType").get_text(strip=True) if record.find("ContractType") else None,
        "ContractTime": record.find("ContractTime").get_text(strip=True) if record.find("ContractTime") else None,
        "Category": record.find("Category").get_text(strip=True) if record.find("Category") else None,
        "Salary": record.find("Salary").get_text(strip=True) if record.find("Salary") else None,
        "OpenDate": record.find("OpenDate").get_text(strip=True) if record.find("OpenDate") else None,
        "CloseDate": record.find("CloseDate").get_text(strip=True) if record.find("CloseDate") else None,
        "SourceName": current_source,
    }
    data.append(record_data)
df = pd.DataFrame(data)
df.info()
```

```python
import pandas as pd
from datetime import datetime

# Function to format dates
def format_date(date_string):
    if date_string:
        try:
            # Try parsing with yyyyMMddTHHmmss format
            parsed_date = datetime.strptime(date_string, "%Y%m%dT%H%M%S")
        except ValueError:
            try:
                # If that fails, try parsing with yyyyddMMTHHmmss format
                parsed_date = datetime.strptime(date_string, "%Y%d%mT%H%M%S")
            except ValueError:
                return None
        # Output always in the y-m-d format
        formatted_date = parsed_date.strftime("%Y-%m-%d %H:%M:%S")
        return formatted_date
    else:
        return None

# Apply the format_date function to the DataFrame columns
df["OpenDate"] = df["OpenDate"].apply(format_date)
df["CloseDate"] = df["CloseDate"].apply(format_date)

# Add an entry to the errors DataFrame
new_entry = {
    'indexOfdf': "ALL",
    'Id': "ALL",
    'ColumnName': ["OpenDate", "CloseDate"],
    'Original': "ALL",
    'Modified': "ALL",
    'ErrorType': 'Format',
    'Fixing': "Changed to YYYY-MM-DD HH:MM:SS format",
}

# Ensure errors DataFrame is defined, if not, create it
if 'errors' not in globals():
    errors = pd.DataFrame(columns=['indexOfdf', 'Id', 'ColumnName', 'Original', 'Modified', 'ErrorType', 'Fixing'])

# Add the new entry using pd.concat
errors = pd.concat([errors, pd.DataFrame([new_entry])], ignore_index=True)

# Display the DataFrame
df
```

```python
import pandas as pd

# Define a dictionary to map misspelled locations to correct ones
corrections = {
    "Cembridge": "Cambridge",
    "ABERDEEN": "Aberdeen",
    "Hampshira": "Hampshire",
    "BRISTOL": "Bristol",
    "Livepool": "Liverpool",
    "Oxfords": "Oxford",
    "LANCASHIRE": "Lancashire",
    "SURREY": "Surrey",
    "Leads": "Leeds",
    "DONCASTER": "Doncaster",
    "Londn": "London",
    "Warringtn": "Warrington",
    "birmingham": "Birmingham",
    "london": "London",
    "HAMpshire": "Hampshire",
    "Manchaster": "Manchester",
    "Croydan": "Croydon"
}

# Apply the corrections to the Location column
for index, row in df.iterrows():
    for incorrect, correct in corrections.items():
        if row["Location"] == incorrect:
            df.at[index, "Location"] = correct

            new_entry = {
                'indexOfdf': index,
                'Id': row['Id'],
                'ColumnName': "Location",
                'Original': incorrect,
                'Modified': correct,
                'ErrorType': 'Misspelling',
                'Fixing': "Changed to " + correct,
            }
            # Ensure errors DataFrame is defined, if not, create it
            if 'errors' not in globals():
                errors = pd.DataFrame(columns=['indexOfdf', 'Id', 'ColumnName', 'Original', 'Modified', 'ErrorType', 'Fixing'])

            # Add the new entry using pd.concat
            errors = pd.concat([errors, pd.DataFrame([new_entry])], ignore_index=True)
            break
            
print(df["Location"].value_counts())
```

```python
# Iterate over DataFrame rows to handle missing values and log changes
for index, row in df.iterrows():
    original_salary = row["Salary"]
    
    # Handle missing values
    if pd.isna(original_salary) or original_salary in ["-","  "," ","","N/A", "None"]:
        corrected_salary = np.nan  # Set the value to NaN
        if original_salary != corrected_salary:
            df.at[index, "Salary"] = corrected_salary
            new_entry = {
                'indexOfdf': index,
                'Id': row['Id'],
                'ColumnName': "Salary",
                'Original': original_salary,
                'Modified': "nan",
                'ErrorType': 'Missing Value',
                'Fixing': "Changed to nan",
            }
            errors = pd.concat([errors, pd.DataFrame([new_entry])], ignore_index=True)

#Correcting other strings
corrections = {
    r"\d*\s(to)\s": "",
    r"\d*\s*(-)\s*": "",
    r"\d*\s*(~)\s*": "",
    r"\d*\s\s": "",
    r"/year": "",
    r"\sper\sAnnum": "",
    r"k": "000"
}

# Apply corrections and log errors
for index, row in df.iterrows():
    original_salary = row["Salary"]
    corrected_salary = original_salary
    for pattern, replacement in corrections.items():
        corrected_salary = re.sub(pattern, replacement, str(corrected_salary))
    
    if original_salary != corrected_salary:
        df.at[index, "Salary"] = corrected_salary
        new_entry = {
            'indexOfdf': index,
            'Id': row['Id'],
            'ColumnName': "Salary",
            'Original': original_salary,
            'Modified': corrected_salary,
            'ErrorType': 'Irregularity',
            'Fixing': f"Changed to {corrected_salary}",
        }
        errors = pd.concat([errors, pd.DataFrame([new_entry])], ignore_index=True)


# # Convert hourly salaries to annual, convert column to float
def convert_to_annual(salary):
    if isinstance(salary, float):
        return salary
    suffixes = [" per hour"," p/h"]
    for suffix in suffixes:
        if salary and salary.endswith(suffix):
            hourly_rate = float(salary.replace(suffix, ""))
            weekly_rate = hourly_rate * 40  # Assuming 40 hours per week
            annual_rate = weekly_rate * 52  # Assuming 52 weeks per year
            return annual_rate
    return salary

# Iterate over DataFrame rows to apply salary conversion and log changes
for index, row in df.iterrows():
    original_salary = row["Salary"]
    corrected_salary = convert_to_annual(original_salary)  # Apply conversion function
    if original_salary != corrected_salary:
        df.at[index, "Salary"] = corrected_salary
        new_entry = {
            'indexOfdf': index,
            'Id': row['Id'],
            'ColumnName': "Salary",
            'Original': original_salary,
            'Modified': corrected_salary,
            'ErrorType': 'Irregularity',
            'Fixing': f"Changed to {corrected_salary}",
        }
        errors = pd.concat([errors, pd.DataFrame([new_entry])], ignore_index=True)

# convert to float
df["Salary"] = df["Salary"].astype(float).round(2)
df.info()
```

```python
#Outliers
sns.boxplot(data=df, y="Salary")
plt.title(f"Box Plot of Salary")
plt.ylabel("Salary")
plt.show()

#show outliers
outliers = df[df["Salary"] > 1000000]
print(outliers)

# Replace outliers with category mean
for index, row in outliers.iterrows():
    category = row['Category']
    category_mean = df[df['Category'] == category]['Salary'].mean()
    df.at[index, 'Salary'] = category_mean

    # Log outliers in error list
    new_entry = {
        'indexOfdf': str(index),
        'Id': row['Id'],
        'ColumnName': 'Salary',
        'Original': str(row['Salary']),
        'Modified': str(category_mean),
        'ErrorType': 'Outlier',
        'Fixing': f"Replaced with category mean ({category})",
    }
    errors = pd.concat([errors, pd.DataFrame([new_entry])], ignore_index=True)

sns.boxplot(data=df, y="Salary")
plt.title(f"Box Plot of Salary")
plt.ylabel("Salary")
plt.show()
```

```python
# Integrity Constraint
# Check for closing dates happening before opening dates
invalid_rows = df[df["CloseDate"] < df["OpenDate"]]
invalid_rows
```

```python
# switch if open date is after closed date
for index, row in invalid_rows.iterrows():
    if row["OpenDate"] > row["CloseDate"]:
        df.at[index, "OpenDate"], df.at[index, "CloseDate"] = row["CloseDate"], row["OpenDate"]
        
new_entry = {
        'indexOfdf': ["18334","30060","34230","36872","48898"],
        'Id': ["68694620","52269401","71741033","68496211","68846246"],
        'ColumnName': ["OpenDate","CloseDate"],
        'Original': "ALL",
        'Modified': "ALL",
        'ErrorType': 'Integrity Constraint',
        'Fixing': "Switched OpenDate and CloseDate",
    }
errors = pd.concat([errors, pd.DataFrame([new_entry])], ignore_index=True)
```
